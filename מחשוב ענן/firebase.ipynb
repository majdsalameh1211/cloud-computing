```python
from firebase import firebase
FBconn = firebase.FirebaseApplication('https://firstdb-6bc58-default-rtdb.asia-southeast1.firebasedatabase.app/',None)
while True:
    Name = str (input ("what is the Name?"))
    LastName = str (input ("what is the Last Name?"))
    Email = str (input ("what is the Email?"))
    Courses = str (input ("what is the Courses?"))
    data_to_upload = {
        'Name' : Name ,
        'Last Name' : LastName ,
        'Email' : Email ,
        'Courses' : Courses ,
    }
    result = FBconn.post('/myTest1/',data_to_upload)

    result = FBconn.get('/myTest1', None)
    if result: # Check if result is not empty
        last_key = list(result.keys())[-1] # Get the key of the last entry
        last_entry = result[last_key]
        print(last_entry) # Print the last entry
    else:
        print("No data in the database.")

    if(Name ==" "):
      break




```
